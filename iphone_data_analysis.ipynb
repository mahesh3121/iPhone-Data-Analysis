```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df1=pd.read_csv('apple_products.csv')
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
highest_rated=df1.sort_values(by=['Star Rating'],ascending=False)
highest_rated=highest_rated.head(10)
print(highest_rated['Product Name'])
```

**Number of Reviews the Highest Rated iphones**

```python
iphones=highest_rated['Product Name'].value_counts()
label=iphones.index
counts=highest_rated['Number Of Ratings']
fig=px.bar(highest_rated,x=label,y=counts,title="Number of Ratings of Highest Rated iPhones")
fig.show()
```

```python
highest_rated['Product Name']
```

```python
pd.crosstab(highest_rated['Product Name'],highest_rated['Discount Percentage'])
```

```python
iphones=highest_rated['Product Name'].value_counts()
label=iphones.index
counts=highest_rated['Number Of Reviews']
fig2=px.bar(highest_rated,x=label,y=counts,title="Number of reviews of highets rated iphones")
fig2.show()
```

**Relationship between the sale price of iPhones and their ratings on Flipkart**

```python
fig=px.scatter(data_frame=df1,x="Number Of Ratings",y="Sale Price",size="Discount Percentage",trendline='ols',title="Relationship between the sale price of iPhones and their ratings on Flipkart")
fig.show()
```

```python
fig2=px.scatter(data_frame=df1,x="Number Of Ratings",y="Discount Percentage",size="Sale Price",trendline="ols",title="Relation Between Number of Rating and Discount")
fig2.show()
```
